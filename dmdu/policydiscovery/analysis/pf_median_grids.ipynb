# Differences between the KPI medians of problem formulations

```python
from dmdu.general.visualization import plot_pf_median_relations
import matplotlib.pyplot as plt
from model.enumerations import ProblemFormulation
import os
import pandas as pd
from ema_workbench import load_results
from dmdu.general.xlm_constants_epsilons import get_all_kpi_names
```

```python

# Directory
target_directory = os.path.join(
    os.path.dirname(os.getcwd()),
    'data',
    'experiments_bad'
)
```

```python
problem_formulations = [
    ProblemFormulation.UTILITARIAN_AGGREGATED,
    ProblemFormulation.UTILITARIAN_DISAGGREGATED,
    ProblemFormulation.SUFFICIENTARIAN_AGGREGATED,
    ProblemFormulation.SUFFICIENTARIAN_DISAGGREGATED,
    ProblemFormulation.EGALITARIAN_AGGREGATED,
    ProblemFormulation.EGALITARIAN_DISAGGREGATED,
    ProblemFormulation.PRIORITARIAN_AGGREGATED,
    ProblemFormulation.PRIORITARIAN_DISAGGREGATED
]

n_references = 4
n_seeds = 2

problem_formulations_dict = {}
## Load results
for problem_formulation in problem_formulations:
    # Define path to load data from
    file = f'results_{problem_formulation.name}'
    path = os.path.join(target_directory, file)

    # Load results
    _, outcomes = load_results(path)
    outcomes = pd.DataFrame(outcomes)

    problem_formulations_dict[problem_formulation.name] = outcomes
```

```python
plt.plot()
```

```python
kpi_list = {
    "Utility": True,
    "Disutility": False,
    "Intratemporal consumption Gini": False,
    "Intratemporal damage Gini": False,
    "Lowest income per capita": True,
    "Highest damage per capita": False,
    "Distance to consumption threshold": False,
    "Population below consumption threshold": False,
    "Distance to damage threshold": False,
    "Population above damage threshold": False,
    "Temperature overshoot": False,
    'Damages': False,
    'Industrial Emission': False,
    'Atmospheric Temperature': False,
    'Total Output': True
}
year = 2105
```

```python
for kpi, direction in kpi_list.items():
    plot_pf_median_relations(
        problem_formulations_dict=problem_formulations_dict,
        kpi=kpi,
        year=year,
        maximize=direction,
        saving=True,
        # file_name=f'kpi_{kpi}_{year}'
)
```
