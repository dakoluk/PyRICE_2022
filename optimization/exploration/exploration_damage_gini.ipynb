## Determine weird cases

```python
length = len(outcomes.columns)
weird_cases = []
non_weird_cases = []

for row_idx, row in outcomes.iterrows():
    for col_idx in range(length):
        if col_idx < length -1:
            case = row_idx
            if abs(row.iloc[col_idx] - row.iloc[col_idx+1]) > 0.1:
                weird_cases.append(case) if case not in weird_cases else 0

for case in range(len(outcomes)):
    if case not in weird_cases:
        non_weird_cases.append(case)
```

```python
weird_outcomes = outcomes.iloc[weird_cases, :]
print(len(weird_outcomes))
plot_one_pathway(experiments, weird_outcomes, 'Intratemporal damage Gini')
```

```python
non_weird_outcomes = outcomes.iloc[non_weird_cases, :]
print(len(non_weird_outcomes))
plot_one_pathway(experiments, non_weird_outcomes, 'Intratemporal damage Gini')
```
